```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Data Preprocessing

```
# use url as the file path
fp = 'https://raw.githubusercontent.com/toliyang/AI_ML/refs/heads/main/Association%20rule%20learning/S_28/Market_Basket_Optimisation.csv'

# get the dataset
dataset = pd.read_csv(fp, header = None)
# get no of rows/cols
rows = dataset.shape[0]
cols = dataset.shape[1]

# new implemenation to parse the text by row
transactions = []
for i in dataset.index:
  val=[str(item) for item in dataset.values[i] if item is not np.nan]
  #val=[str(item) for item in dataset.values[i]]
  transactions.append(val)

# create data list
#transactions = []
#for i in range(0, rows):
#  transactions.append([str(dataset.values[i,j]) for j in range(0, cols)])
```

## Training the Apriori model on the dataset

```
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 3)
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```
results = list(rules)
```

```
results
```

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```
resultsinDataFrame.nlargest(n = 10, columns = 'Lift')
```
