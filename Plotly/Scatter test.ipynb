```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python



# Create data
x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)
X, Y = np.meshgrid(x, y)
Z = X**2*Y+np.sin(X) # The function z = 10x + y

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])

# Update layout for better viewing
fig.update_layout(
    title='3D Plot of z = 10x + y',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        camera=dict(eye=dict(x=1.5, y=1.5, z=0.5))  # Adjust view angle
    ),
    autosize=True,
    width=800,
    height=600
)

fig.show()
```
